```python
df = pd.read_csv("/home/m3m/INFO_PROYECTO/drexml_paper_paquete/my_analysis.sig_queryl1k_tool.65d76720ec82c00013c4cb79/arfs/TAG/query_result.gct", skiprows=2, sep='\t', index_col=0)
```

```python
df = df.drop("desc")
```

```python
df
```

```python
df["norm_cs"] = df["norm_cs"].astype(float, errors="ignore")
```

```python
## si se quieren incluir las silenciaciones génicas como targets le pongo el gen silenciado a la columna de targets
df.reset_index(inplace=True)
df.loc[df[(df["pert_type"] != "trt_cp") & (df["pert_type"] != "trt_lig")].index.unique(), "target_name"] = df["pert_iname"]
df.set_index("id", inplace=True)
```

```python
if only_compounds:
    df = df[(df["pert_type"] == "trt_cp") | (df["pert_type"] == "trt_lig")]
```

```python
comp = df[(df["target_name"] != "-666")]
```

```python
comp["norm_cs"] = comp["norm_cs"].astype(float)
```

```python
## la columna de targets es un string en el que los diferentes targets de un compuesto están separados por un "|"
## hago un split y merge para tener una fila por target
split_comp = comp["target_name"].str.split("|", expand=True).stack().reset_index(level=1, drop=True).to_frame('single_target')
comp = comp.drop("target_name", axis=1).merge(split_comp, left_index=True, right_index=True)
```

```python
## ordenar por mayor valor absoluto y quedarse con el mayor valor absoluto
comp = comp.reset_index()
comp = comp.reindex(comp["norm_cs"].abs().sort_values(ascending=False).index)
comp = comp.set_index("id")
comp = comp.drop_duplicates(subset=["single_target"], keep="first")
```

```python
top5 = comp.nlargest(int(0.05*comp.single_target.nunique()), "norm_cs")
bo5 = comp.nsmallest(int(0.05*comp.single_target.nunique()), "norm_cs")
```

```python
comp
```

```python
most_rel = pd.concat([top5, bo5])
```

```python
## los targets escogidos por ambos métodos
inter = most_rel[most_rel["single_target"].isin(sel_targets.columns)]
```
